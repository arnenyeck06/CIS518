```python
df2 = df1.drop(['petal_width'],axis = 1)
df2
```

# Label Encoder to encode the class target variables to numeric values.

```python
##Encode categorical values to numbers
label_encoder = LabelEncoder()
df2.iloc[:,-1] = label_encoder.fit_transform(df2.iloc[:,-1]).astype('float64')
df2
```

```python
df2['Class'].unique() ##Gives list of distinct variables present in 'Class' Columns
```

# Model Evaluation

```python
# Split dataset into features and target
y = df2['Class'] ##Target Variable
X= df2[['sepal_length','sepal_width','petal_length']]
```

```python
X
```

```python
# Split features and target into train and test sets into (70-30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, stratify = y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# Logistic Regression Model

```python
# Instantiate and fit the LR
Model = LogisticRegression()
Model.fit(X_train, y_train)

y_pred = Model.predict(X_test)
```

```python
petal=df1[['petal_length','petal_width','Class']] #high correlated features
sepal=df1[['sepal_length','sepal_width','Class']] #low correlated features



train_p,test_p=train_test_split(petal,test_size=0.3,random_state=0) #petals
train_x_p=train_p[['petal_width','petal_length']]
train_y_p=train_p.Class
test_x_p=test_p[['petal_width','petal_length']]
test_y_p=test_p.Class




train_s,test_s=train_test_split(sepal,test_size=0.3,random_state=0) #Sepal
train_x_s=train_s[['sepal_width','sepal_length']]
train_y_s=train_s.Class
test_x_s=test_s[['sepal_width','sepal_length']]
test_y_s=test_s.Class



model = LogisticRegression()
model.fit(train_x_p,train_y_p)
prediction=model.predict(test_x_p)
print('The accuracy of the Logistic Regression using Petals is:',metrics.accuracy_score(prediction,test_y_p))



model.fit(train_x_s,train_y_s)
prediction=model.predict(test_x_s)
print('The accuracy of the Logistic Regression using Sepals is:',metrics.accuracy_score(prediction,test_y_s))
```

```python

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
print('accuracy is',accuracy_score(y_pred,y_test))
```

# Hyper-parameter Tuning

```python
#create a logistic regression object with an l2 penalty
logistic = linear_model.LogisticRegression()
```

```python
## Create Dictionary
param_grid = [    
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-1, 1, 3),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [100,200,1000]
    }
]
```

```python
lr_gr = GridSearchCV(logistic,param_grid = param_grid,cv=2,verbose = True,n_jobs = -1) #Use GridSearchCV 
```

```python
 best_lr_gr=lr_gr.fit(X_train,y_train) ##Fit to X_train & y_train data
```

```python
best_lr_gr.best_estimator_ #Selects the best estimator
```

```python
print (f'Accuracy - : {best_lr_gr.score(X_train,y_train):.3f}') ## Checks for Accuracy score
```

```python
# Evaluate the model
print("Training scores: {:.2f}".format(best_lr_gr.score(X_train, y_train)))
print("Test scores: {:.2f}".format(best_lr_gr.score(X_test,y_test)))
```
